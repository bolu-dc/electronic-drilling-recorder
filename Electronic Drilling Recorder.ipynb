```python
# Load the Data
import pandas as pd
drilling_data = pd.read_csv('EDR_data.csv', sep=',', header=0)
drilling_data['Time'] = pd.to_datetime(drilling_data['Time'])

drilling_data.head()
```

```python
import matplotlib.pyplot as plt
import datetime
```

```python
# Get the time interval
def find_time_array(start_time, end_time):
  time_array = []
  start_time_list = str(start_time).split(":")
  start_time_hour = int(start_time_list[0])
  start_time_min = int(start_time_list[1])
  start_time_sec = int(start_time_list[2])
  start_time = datetime.time(start_time_hour, start_time_min, start_time_sec)

  end_time_list = str(end_time).split(":")
  end_time_hour = int(end_time_list[0])
  end_time_min = int(end_time_list[1])
  end_time_sec = int(end_time_list[2])
  end_time = datetime.time(end_time_hour, end_time_min, end_time_sec)

  for i in drilling_data['Time']:
    if i.time()<=end_time and i.time()>= start_time:
      time_array.append(i)
  return time_array

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
plt.figure(figsize=(30,10))
fig.suptitle('EDR_Plots')

# Subplot 1
start_time1 = "0:00:00"
end_time1 = "6:39:24"
time_array1 = find_time_array(start_time1, end_time1)
temp_df1 = drilling_data[drilling_data['Time'].isin(time_array1)]
ax1.plot(temp_df1['Bit Depth (feet)'], temp_df1['Time'], label = "Bit Depth (feet)")
ax1.plot(temp_df1['Hole Depth (feet)'], temp_df1['Time'], label = "Hole Depth (feet)")
ax1.legend(loc="upper right")

# Subplot 2
start_time2 = "0:00:00"
end_time2 = "6:39:24"
time_array2 = find_time_array(start_time2, end_time2)
temp_df2 = drilling_data[drilling_data['Time'].isin(time_array2)]
ax2.plot(temp_df2['Rate Of Penetration (ft_per_hr)'], temp_df2['Time'], label = "Rate Of Penetration (ft_per_hr)")
ax2.plot(temp_df2['Weight on Bit (klbs)'], temp_df2['Time'], label = "Weight on Bit (klbs)")
ax2.plot(temp_df2['Rotary RPM (RPM)'], temp_df2['Time'], label = "Rotary RPM (RPM)")
ax2.legend(loc="upper right")

# Subplot 3
start_time3 = "0:00:00"
end_time3 = "6:39:24"
time_array3 = find_time_array(start_time3, end_time3)
temp_df3 = drilling_data[drilling_data['Time'].isin(time_array3)]
ax3.plot(temp_df3['Standpipe Pressure (psi)'], temp_df3['Time'], label = "Standpipe Pressure (psi)")
ax3.plot(temp_df3['Hook Load (klbs)'], temp_df3['Time'], label = "Hook Load (klbs)")
ax3.plot(temp_df3['Differential Pressure (psi)'], temp_df3['Time'], label = "Differential Pressure (psi)")
ax3.legend(loc="upper right")

# Subplot 4
start_time4 = "0:00:00"
end_time4 = "6:39:24"
time_array4 = find_time_array(start_time4, end_time4)
temp_df4 = drilling_data[drilling_data['Time'].isin(time_array4)]
ax4.plot(temp_df4['Flow (flow_percent)'], temp_df4['Time'], label = "Flow (flow_percent)")
ax4.plot(temp_df4['Total Pump Output (gal_per_min)'], temp_df4['Time'], label = "Total Pump Output (gal_per_min)")
ax4.plot(temp_df4['Block Height (feet)'], temp_df4['Time'], label = "Block Height (feet)")
ax4.legend(loc="upper right")

# Show the plot
plt.show()
```
